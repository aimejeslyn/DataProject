**MODEL**

```r
sort(colnames(df2))

colnames(df2) <- c('Daily_steps','Heart_rate','Physical_activity_level','Quality_of_sleep','Sleep_duration','Stress_level')
sort(colnames(df2))
```

```r
set.seed(1)

#use 70% of dataset as training set and 30% as test set
sample <- sample(c(TRUE, FALSE), nrow(df2), replace=TRUE, prob=c(0.7,0.3))
train  <- df2[sample, ]
test   <- df2[!sample, ]
```

```r
y_train <- train[c('Quality_of_sleep')]
y_test <- test[c('Quality_of_sleep')]

# get independent variables
x_train <- train[c('Daily_steps','Heart_rate','Physical_activity_level','Sleep_duration','Stress_level')]
x_test <- test[c('Daily_steps','Heart_rate','Physical_activity_level','Sleep_duration','Stress_level')]


# validasi hasil split
cat("\njumlah baris x train:", nrow(x_train))
cat("\njumlah baris x test:", nrow(x_test))

cat("\njumlah kolom x: ", ncol(x_train))
cat("\njumlah kolom y: ", ncol(y_train))
```

```r
#build model
model1 <- lm(Quality_of_sleep ~ Daily_steps + Heart_rate + Physical_activity_level + Sleep_duration + Stress_level , data = train)
summary(model1)
```

```r
install.packages('Metrics')
```

```r
library(Metrics)

test$pred_sleepquality1 <- predict(model1, newdata=x_test)

cat('\nMAE: ', mae(y_test$Quality_of_sleep, test$pred_sleepquality1))
cat('\nMSE: ', mse(y_test$Quality_of_sleep, test$pred_sleepquality1))
cat('\nRMSE: ',rmse(y_test$Quality_of_sleep, test$pred_sleepquality1))
```

```r
ggplot(test,aes(Quality_of_sleep,pred_sleepquality1)) +
    geom_point(color="red") +
    geom_smooth(method="lm",color="red",size=1) +
    ggtitle("Predicted Sleep Quality vs Actual Sleep Quality") + theme(plot.title = element_text(hjust = 0.5)) +
    xlab("Actual Sleep Quality") +
    ylab("Predicted Sleep Quality")
```

```r
model2 <- lm(Quality_of_sleep ~ Physical_activity_level + Sleep_duration , data = train)

test$pred_sleepquality2 <- predict(model2, newdata=x_test)

cat('\nMAE: ', mae(y_test$Quality_of_sleep, test$pred_sleepquality2))
cat('\nMSE: ', mse(y_test$Quality_of_sleep, test$pred_sleepquality2))
cat('\nRMSE: ',rmse(y_test$Quality_of_sleep, test$pred_sleepquality2))
```

```r
ggplot(test,aes(Quality_of_sleep,pred_sleepquality2)) +
    geom_point(color="red") +
    geom_smooth(method="lm",color="red",size=1) +
    ggtitle("Predicted Sleep Quality vs Actual Sleep Quality") + theme(plot.title = element_text(hjust = 0.5)) +
    xlab("Actual Sleep Quality") +
    ylab("Predicted Sleep Quality")
```
